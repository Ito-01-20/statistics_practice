## 装飾

```python
x = np.linspace(-10, 10, 10)
z = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x, z, 'g--', linewidth=5)
```

```python
plt.plot(x, z, color='r', 
         lw=3, 
         ls ='--', 
         marker='o', 
         markersize=10, 
         markerfacecolor='yellow',
        alpha=0.5)
```

## 付属情報をつける

```python
plt.plot(x, z)
plt.plot(x, z*2, label='sigmoid * 2')
plt.xlabel('x')
plt.xlabel('y')
plt.title('sigmoid(x)')
plt.legend(loc=4)
plt.xticks([1, 2, 3], rotation=20)
plt.xticks([0, 0.5, 1], rotation=20)
```

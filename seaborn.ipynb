## Pandas

```python
import pandas as pd
import numpy as np
```

### series

```python
data = {
    'Name': 'john',
    'Sex': 'male',
    'Age': 22
}
```

```python
pd.Series(data)
```

```python
array = np.array([22, 31, 42, 23])
pd.Series(array)
```

```python
array = np.array(['john', 'male', 22])
pd.Series(array, index=['Name', 'Sex', 'Age'])
```
